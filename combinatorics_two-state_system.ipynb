### Calculating and plotting two-state probabilities

We'll use $N$ to denote the total number of two-state units (coins or paramagnets or $\ldots$) and $n$ to be the number of heads, spin up, or $\ldots$

The total number of states is $2^N$ and the multiplicity of $n$ is
$$
 \Omega(N,n) = \frac{N!}{n! (N-n)!} = {N\choose n} \,.
$$

```python
import numpy as np
from scipy.special import comb
import matplotlib.pyplot as plt
```

```python
def prob(n, N=50):
    """
    Calculate the probability of n heads when tossing N fair coins.
    This maps onto any two-state system.
    """
    return comb(N, n) / 2**N
```

```python
prob(25, 50)
```

```python
N_test = 50
n_array = np.arange(N_test)
plt.scatter(n_array, prob(n_array, N_test), )
plt.xlabel("# of heads")
plt.ylabel("probability");
```

```python
def plotit(ax, N=50):
    """
    Make a plot of the probabilities from 0 to N. 
    """
    n_array = np.arange(N)
    ax.scatter(n_array, prob(n_array, N_test))
    ax.set_xlabel("# of heads")
    ax.set_ylabel("probability")
    
    return ax
```

```python
fig, (ax0) = plt.subplots(1, 1, constrained_layout=True)
plotit(ax0, N_test)
fig.tight_layout()
```

# Expansion of $a^x$

When $x\ln a \ll 1$,

$$
  a^x = e^{x\ln a} \approx 1 + x \ln a +  \frac12 x^2 (\ln a)^2 + \cdots
$$

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define functions

```python
def full(x, a=2):
    """
    f[x] = a^x, with a=2 by default
    """
    return a**x

def first_order(x, a=2):
    """
    First order expansion of a^x --> 1 + x ln a
    """
    return 1 + x * np.log(a)


def second_order(x, a=2):
    """
    Second order expansion of a^x --> 1 + x ln a + 1/2 x^2 (ln a)^2
    """
    return 1 + x * np.log(a) + 1/2 * x**2 * (np.log(a))**2
```

## Make some plots

```python
# Set the global font size to 14 (points)
plt.rcParams['font.size'] = 14
```

```python
a = 2
x = np.linspace(0, 1, 200)
```

```python
fig, ax = plt.subplots(1, 1, constrained_layout=True)
ax.plot(x, full(x, a), label=r'$a^x$', color='black')
ax.plot(x, first_order(x, a), label=r'first order', color='blue')
ax.plot(x, second_order(x, a), label=r'second order', color='red')
ax.set_title(r'$a^x$ and expansions')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$f(x)$')
ax.legend()
plt.show()
```

## Stirling's approximation plots

$$ \ln N! \approx N \ln N - N + \ln(\sqrt{2\pi N})  \qquad \text{for }N\gg1 
$$
or
$$  N! \approx N^N e^{-N} \sqrt{2\pi N}
$$

```python
from scipy.special import gammaln

def log_fact(n):
    """
    Calculate the log factorial using a defined logarithm of the gamma function
    plus n! = \Gamma(n+1)
    """
    return gammaln(n + 1)
```

```python
n_min = 10
n_max = 100
n_array = np.arange(n_min, n_max)
```

```python
fig, ax = plt.subplots(1, 1, constrained_layout=True)
ax.plot(n_array, log_fact(n_array), label=r'$\ln N!$', color='black')
ax.plot(n_array, n_array * np.log(n_array) - n_array, label=r'leading order', color='blue')
# ax.plot(x, second_order(x, a), label=r'second order', color='red')
# ax.set_title(r'$a^x$ and expansions')
ax.set_xlabel(r'$N$')
ax.set_ylabel(r'$\ln N!$ and Stirling approximations')
ax.legend()
plt.show()
```
